## Exercise 3.02: Creating a Pie Chart for Water Usage

```python
# Import statements
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

Use pandas to read the data located in the subfolder data.

```python
water_usage = 'https://raw.githubusercontent.com/sandeep92134/The-Data-Visualization-Workshop-by-packt/master/module%203/datasets/water_usage.csv'
```

```python
# Load dataset
data = pd.read_csv(water_usage)
```

Use a pie chart to visualize the water usage. Highlight one usage of your choice using the explode parameter. Show the percentages for each slice and add a title.

```python
# Create figure
plt.figure(figsize=(5, 5), dpi=300)
# Create pie plot
plt.pie('Percentage', explode=(0, 0, 0.1, 0, 0, 0), labels='Usage', data=data, autopct='%.0f%%')
# Add title
plt.title('Water usage')
# Show plot
plt.show()
```
