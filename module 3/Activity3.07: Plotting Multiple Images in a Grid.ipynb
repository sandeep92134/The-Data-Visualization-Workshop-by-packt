## Activity 3.07: Plotting Multiple Images in a Grid

In this activity, we will plot images in a grid.

```python
# Import statements
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

%matplotlib inline
```

Load all four images from the subfolder data.

```python
# Load images
img_filenames = sorted(os.listdir('/content/images'))
imgs = [mpimg.imread(os.path.join('/content/images', img_filename)) for img_filename in img_filenames]
```

Visualize the images in a 2x2 grid. Remove the axes and give each image a label.

```python
# Create subplot
fig, axes = plt.subplots(2, 2)
fig.figsize = (6, 6)
fig.dpi = 200
axes = axes.ravel()
# Specify labels
labels = ['coast', 'beach', 'building', 'city at night']
# Plot images
for i in range(len(imgs)):
    axes[i].imshow(imgs[i])
    axes[i].set_xticks([])
    axes[i].set_yticks([])
    axes[i].set_xlabel(labels[i])
plt.show()
```
