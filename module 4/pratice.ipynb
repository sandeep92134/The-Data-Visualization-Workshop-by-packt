```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Advantages of Seaborn

```python
sns.set(style="ticks")
data = pd.read_csv("https://raw.githubusercontent.com/sandeep92134/The-Data-Visualization-Workshop-by-packt/master/module%204/datasets/salary.csv")[:1000]
plt.figure()
sns.relplot(x="Salary", y="Age", hue="Education", style="Education",
col="Gender", data=data)
plt.show()
```

## Controlling Figure Aesthetics

```python
plt.figure()
x1 = [10, 20, 5, 40, 8]
x2 = [30, 43, 9, 7, 20]
plt.plot(x1, label='Group A')
plt.plot(x2, label='Group B')
plt.legend()
plt.show()
```

```python
sns.set()
plt.figure()
x1 = [10, 20, 5, 40, 8]
x2 = [30, 43, 9, 7, 20]
plt.plot(x1, label='Group A')
plt.plot(x2, label='Group B')
plt.legend()
plt.show()
```

## Seaborn Figure Styles

```python
sns.set_style("whitegrid")
plt.figure()
x1 = [10, 20, 5, 40, 8]
x2 = [30, 43, 9, 7, 20]
plt.plot(x1, label='Group A')
plt.plot(x2, label='Group B')
plt.legend()
plt.show()
```

```python
sns.set()
plt.figure()
x1 = [10, 20, 5, 40, 8]
x2 = [30, 43, 9, 7, 20]
with sns.axes_style('dark'):
    plt.plot(x1, label='Group A')
    plt.plot(x2, label='Group B')
plt.legend()
plt.show()
```

## Removing Axes Spines

```python
sns.set_style("white")
plt.figure()
x1 = [10, 20, 5, 40, 8]
x2 = [30, 43, 9, 7, 20]
plt.plot(x1, label='Group A')
plt.plot(x2, label='Group B')
sns.despine()
plt.legend()
plt.show()
```

## Controlling the Scale of Plot Elements

```python
sns.set_context("poster")
plt.figure()
x1 = [10, 20, 5, 40, 8]
x2 = [30, 43, 9, 7, 20]
plt.plot(x1, label='Group A')
plt.plot(x2, label='Group B')
plt.legend()
plt.show()
```

## Bar Plots

```python
data = pd.read_csv("https://raw.githubusercontent.com/sandeep92134/The-Data-Visualization-Workshop-by-packt/master/module%204/datasets/salary.csv")
sns.set(style="whitegrid")
plt.figure()
sns.barplot(x="Education", y="Salary", hue="District", data=data)
plt.show()
```

## Kernel Density Estimation

```python
data = pd.read_csv('https://raw.githubusercontent.com/sandeep92134/The-Data-Visualization-Workshop-by-packt/master/module%204/datasets/age_salary_hours.csv')
```

```python
sns.set_style('white')
sns.distplot(data.loc[:, 'Age'])
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```

```python
sns.kdeplot(data.loc[:, 'Age'], shade=True)
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```

## Plotting Bivariate Distributions

```python
data = pd.read_csv('https://raw.githubusercontent.com/sandeep92134/The-Data-Visualization-Workshop-by-packt/master/module%204/datasets/age_salary_hours.csv')
```

```python
sns.set(style="white")
sns.jointplot(x="Annual Salary", y="Age", data=data)
plt.show()
```

```python
sns.jointplot('Annual Salary', 'Age', data=data, kind='kde', xlim=(0, 500000), ylim=(0, 100))
plt.show()
```

## Visualizing Pairwise Relationships

```python
data = pd.read_csv('https://raw.githubusercontent.com/sandeep92134/The-Data-Visualization-Workshop-by-packt/master/module%204/datasets/age_salary_hours.csv')
```

```python
sns.set(style="ticks", color_codes=True)
g = sns.pairplot(data, hue='Education')
plt.show()
```

## Violin Plots

```python
data = pd.read_csv("https://raw.githubusercontent.com/sandeep92134/The-Data-Visualization-Workshop-by-packt/master/module%204/datasets/salary.csv")
sns.set(style="whitegrid")  
sns.violinplot('Education', 'Salary', hue='Gender', data=data, split=True, cut=0)
plt.show()
```

## FacetGrid

```python
sns.set()
data = pd.read_csv("https://raw.githubusercontent.com/sandeep92134/The-Data-Visualization-Workshop-by-packt/master/module%204/datasets/salary.csv")[:1000]
g = sns.FacetGrid(data, col='District')
g.map(plt.scatter, 'Salary', 'Age')
plt.show()
```

## Regression Plots

```python
sns.set()
x = np.arange(100)
y = x + np.random.normal(0, 5, size=100) # normal distribution with mean 0 and a standard deviation of 5
sns.regplot(x, y)
plt.show()
```

## Squarify

```python
import squarify
colors = sns.light_palette("brown", 4)
squarify.plot(sizes=[50, 25, 10, 15], label=["Group A", "Group B", "Group C", "Group D"], color=colors)
plt.axis("off")
plt.show()
```
