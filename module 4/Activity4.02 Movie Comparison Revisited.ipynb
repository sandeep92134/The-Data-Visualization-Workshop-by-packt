## Activity 4.02: Movie Comparison Revisited
In this activity, we will use a bar plot to compare movie scores. You are given five movies with scores from Rotten Tomatoes. The Tomatometer is the percentage of approved Tomatometer critics who have given a positive review for the movie. The Audience Score is the percentage of users who have given a score of 3.5 or higher out of 5. Compare these two scores among the five movies.

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Use pandas to read the movie_scores.csv data located in the Dataset folder and transform the data into a useable format for Seaborn's barplot function.

```python
mydata = pd.read_csv("https://raw.githubusercontent.com/sandeep92134/The-Data-Visualization-Workshop-by-packt/master/module%204/datasets/movie_scores.csv", index_col=0)
```

```python
mydata
```

```python
movie_scores = pd.DataFrame({"Movie Title": list(mydata["MovieTitle"]) * 2,
                "Score": list(mydata["AudienceScore"]) + list(mydata["Tomatometer"]),
                "Type": ["Audience Score"] * len(mydata["AudienceScore"]) + ["Tomatometer"] * len(mydata["Tomatometer"])})
```

```python
movie_scores
```

Use Seaborn to create a visually-appealing bar plot comparing the two scores for all five movies.

```python
sns.set()
plt.figure(figsize=(8, 5), dpi=300)
# Create bar plot
ax = sns.barplot("Movie Title", "Score", hue="Type", data=movie_scores)
plt.xticks(rotation=10)
# Add title
plt.title("Movies Scores comparison")
plt.xlabel("Movies")
plt.ylabel("Scores")
# Show plot
plt.savefig('seaborn_movie_scores.png', dpi=300, bbox_inches='tight')
```
