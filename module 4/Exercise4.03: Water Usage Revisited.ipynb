```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import squarify
```

Use pandas to read the data located in the Datasets data.

```python
mydata = pd.read_csv("https://raw.githubusercontent.com/sandeep92134/The-Data-Visualization-Workshop-by-packt/master/module%203/datasets/water_usage.csv", index_col=0)
```

```python
mydata
```

Create a list of labels by accessing each column from the preceding dataset. Here, the astype('str') function is used to cast the fetched data into a type string.

```python
labels = mydata['Usage'] + ' (' + mydata['Percentage'].astype('str') + '%)'
```

For creating a tree map visualization of given data, use of the plot() function of the squarify library. This function takes three parameters. The first parameter is the list of all percentages and the second parameter is the list of all labels, which we got in the previous step. The third parameter is the colormap that can be created by using the light_pallete() function of the Seaborn library.

```python
# Create figure
plt.figure(dpi=220)
# Create tree map
squarify.plot(sizes=mydata['Percentage'], label=labels, color=sns.light_palette('green', mydata.shape[0]))
plt.axis('off')
# Add title
plt.title('Water usage')
# Show plot
plt.show()
```
