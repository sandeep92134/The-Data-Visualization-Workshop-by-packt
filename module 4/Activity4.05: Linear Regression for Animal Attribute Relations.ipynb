## Activity 4.05: Linear Regression for Animal Attribute Relations
You have a dataset about various animals including their body mass and maximum longevity. To discover if there is any linear relationship between these two variables a regression plot shall be used.

The following are the steps to perform:

1. Use pandas to read the **anage_data.csv** dataset located in the Datasets folder.
2. Filter the data so that you end up with samples containing a body mass and maximum longevity. Only consider samples for the Mammalia class and a body mass of less than 200,000.
3. Create a regression plot to visualize the linear relationship between the variables.

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mydata = pd.read_csv("https://raw.githubusercontent.com/sandeep92134/The-Data-Visualization-Workshop-by-packt/master/module%204/datasets/anage_data.csv")
```

```python
mydata.head()
```

Filter the data so you end up with samples containing a body mass and a maximum longevity. Only consider samples for the class Mammalia and a body mass below 200,000.

```python
longevity = 'Maximum longevity (yrs)'
mass = 'Body mass (g)'
data = mydata[mydata['Class'] == 'Mammalia']
data = data[np.isfinite(data[longevity]) & np.isfinite(data[mass]) & (data[mass] < 200000)]
```

Create a regression plot to visualize the linear relationship of the variables.

```python
# Create figure
sns.set()
plt.figure(figsize=(8, 6), dpi=300)
# Create scatter plot
sns.regplot(mass, longevity, data=data)
# Show plot
plt.show()
```
